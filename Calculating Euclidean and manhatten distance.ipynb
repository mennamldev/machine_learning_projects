Importing the libraries


```python
import numpy as np
```

```python
# Euclidean Distance
# consider two points with two dimensions
p1 = (1,1)
p2 = (2,2)

```

```python
# calculating Euclidean distance
dist = (p1[0] - p2[0])**2 + (p1[1] - p2[1])**2

euclidean_dist = np.sqrt(dist)
print(euclidean_dist)
```

```python
# Euclidean Distance
# consider two points with three dimensions
p1 = (1,1,1)
p2 = (2,2,2)
```

```python
# calculating Euclidean distance
dist = (p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2]- p2[2])**2

euclidean_dist = np.sqrt(dist)
print(euclidean_dist)
```

```python
 # consider two points with four dimensions
p1 = (1,1,1,1)
p2 = (2,2,2,2)

dist = 0

for i in range(len(p1)):
  dist = dist + (p1[i] - p2[i])**2

  euclidean_dist = np.sqrt(dist)
  print(euclidean_dist)
```

```python
   # creating a function to calculate Euclidean Distance
   def get_euclidean_distance(p1,p2):
    dist = 0

    for i in range(len(p1)):
      dist = dist + (p1[i] - p2[i])**2

    euclidean_dist = np.sqrt(dist)
    print(euclidean_dist)
```

```python
get_euclidean_distance((1,1,1),(2,2,2))
```

```python
# creating a function to calculate Manhattan Distance
def get_Manhattan_distance(p1,p2):
    dist = 0

    for i in range(len(p1)):
      dist = dist + abs(p1[i] - p2[i])
      Manhattan_distance = dist
    print(Manhattan_distance)
```

```python
get_Manhattan_distance((1,1,1,1),(2,2,2,2))
```
