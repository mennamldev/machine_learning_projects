```python
# importing the libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler # to standardize the data
from sklearn.model_selection import train_test_split # to split the data into training and test data
from sklearn import svm # support vector machine to train the model
from sklearn.metrics import accuracy_score # to evaluate the model
```

```python
# Data collection and analysis
# loading the data
diabetes_dataset = pd.read_csv("/content/diabetes.csv")
```

```python
# first five rows
diabetes_dataset.head()
```

```python
# last five rows
diabetes_dataset.tail()
```

```python
# number of rows and columns
diabetes_dataset.shape
```

```python
# getting the statistical measures of the daa
diabetes_dataset.describe()
```

```python
# checking the number of each outcome
diabetes_dataset['Outcome'].value_counts()
```

0 --> non_diabetic

1 --> diabetic

```python
# getting the mean value for each outcome
diabetes_dataset.groupby('Outcome').mean()
```

```python
# seperating the data and labels
x = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
y = diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
# Data standardization to get similar range to get better prediction
scaler = StandardScaler()
```

```python
# fitting the data into the scaler
scaler.fit(x)
```

```python
# transfor the data
standardized_data = scaler.transform(x)
```

```python
print(standardized_data) # we get all the data in the range 0 and 1
```

```python
#rename the values and the labels
x = standardized_data
y = diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
# split the data into training and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2) # sratify means the data will be splited as 1 or 0 like the values in y
```

```python
# checking the shape of the dataset, training_set, test_set
print(x.shape, x_train.shape, x_test.shape)
```

```python
# training the model
classifier = svm.SVC(kernel = 'linear') # svm --> support vector machine, svc --> support vector classifier, kernel = linear --> gives linear model
```

```python
# fit the training data to the svm classifier
classifier.fit(x_train, y_train)
```

```python
# model evaluation, like how manny times the model is predicting correctly
# finding the accuracy score on traing data
x_train_prediction = classifier.predict(x_train) # predict the labels for training set
training_data_accuracy = accuracy_score(x_train_prediction, y_train) # comparing the predicted values to the original ones y_train
```

```python
print('Accuracy score of the training data :', training_data_accuracy)
```

```python
# finding the accuracy score for the test data
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score of the test date :', test_data_accuracy)
```

```python
# making a predictive system
input_data = (4,110,92,0,0,37.6,0.191,30)
# changing the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance in the input data
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# stadardize the input data
std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0 ):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
