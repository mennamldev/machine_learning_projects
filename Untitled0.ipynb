Stemming:

Stemming is the process of reducing a word to it's root word

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmd_content = re.sub("[^a-zA-z]"," ", content)
  stemmd_content = stemmd_content.lower()
  stemmd_content = stemmd_content.split()
  stemmd_content = [port_stem(word) for word in stemmd_content if not word in stopwords.words("english")]
  return stemmd_content
```

```python
news_data["content"] = news_data["content"].apply(stemming)
```

Projecction of A vector by B vector

```python
a = np.array([2, 5])

v = np.array([8, -6])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print(proj_of_a_on_v)
```

```python
a = np.array([23,45,62])

v = np.array([45,82,76])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print(proj_of_a_on_v)
```
